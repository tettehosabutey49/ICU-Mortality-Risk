## sample demo data of only 5,000 entries from our original dataset due to streamlit limitations

```python
import pandas as pd
df = pd.read_csv('data/raw/patient_survival.csv')
sample = df.groupby('hospital_death', group_keys=False).apply(
    lambda x: x.sample(min(len(x), 4500), random_state=42)
)
sample.to_csv('app/data/demo_data.csv', index=False)
```
